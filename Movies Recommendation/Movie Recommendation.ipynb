```python
m1 = pd.merge(r, m, on='movieId')
m1
```

```python
m1.groupby('title')['rating'].mean().head()
```

```python
m1.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
m1.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings_mean = pd.DataFrame(m1.groupby('title')['rating'].mean())
```

```python
ratings_mean['rating_counts'] = pd.DataFrame(m1.groupby('title')['rating'].count())
```

```python
ratings_mean.head()
```

```python
plt.figure(figsize=(10,8))
plt.rcParams['patch.force_edgecolor'] = True
ratings_mean['rating_counts'].hist(bins=50)
```

```python
plt.figure(figsize=(10,8))
plt.rcParams['patch.force_edgecolor'] = True
ratings_mean['rating'].hist(bins=50)
```

```python
user_movie_rating = m1.pivot_table(index='userId', columns='title', values='rating')

user_movie_rating.head()
```

```python
forrest_gump_ratings = user_movie_rating['Forrest Gump (1994)']

forrest_gump_ratings.head()
```

```python
movies_like_forest_gump = user_movie_rating.corrwith(forrest_gump_ratings)
```

```python
corr_forrest_gump = pd.DataFrame(movies_like_forest_gump, columns=['Correlation'])
corr_forrest_gump.dropna(inplace=True)
corr_forrest_gump.head()
```

```python
corr_forrest_gump.sort_values('Correlation', ascending=False).head(10)
```

```python
corr_forrest_gump = corr_forrest_gump.join(ratings_mean['rating_counts'])
corr_forrest_gump.head()
```

```python
corr_forrest_gump[corr_forrest_gump ['rating_counts']>50].sort_values('Correlation', ascending=False).head()
```
